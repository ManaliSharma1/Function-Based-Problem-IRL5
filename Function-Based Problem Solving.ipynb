## Problem 1: Social Media – Post Engagement Analyzer

```python
# Problem 1: Post Engagement Analyzer

def analyze_post_engagement(likes_list):
    """
    Function to calculate total likes and determine post status.
    """
    total_likes = 0
    
    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes
    
    # Conditional check
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Sample Data
likes_data = [200, 300, 150, 400]
analyze_post_engagement(likes_data)
```

## Problem 3: Agriculture – Rainfall Adequacy Checker

```python
# Problem 3: Rainfall Adequacy Checker

def check_rainfall(rainfall_data, required_level):
    """
    Function to calculate average rainfall and check adequacy.
    """
    total = 0
    
    for rain in rainfall_data:
        total += rain
    
    average = total / len(rainfall_data)
    
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)


# Sample Data
rainfall = [70, 75, 80, 65]
check_rainfall(rainfall, 70)
```

## Problem 4: Social Media – Duplicate Account Detection

```python
# Problem 4: Duplicate Account Detection

def detect_duplicate_accounts(usernames):
    """
    Function to detect duplicate usernames using set.
    """
    unique_users = set(usernames)
    
    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Sample Data
users = ["rahul", "sneha", "amit", "rahul"]
detect_duplicate_accounts(users)
```

## Problem 6: Agriculture – Premium Crop Price Filter

```python
# Problem 6: Premium Crop Price Filter

def filter_premium_crops(prices):
    """
    Function to filter crop prices above ₹2000.
    """
    premium = []
    
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    print("Premium Crops:", premium)


# Sample Data
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

## Problem 7: System Monitoring – Application Health Checker

```python
# Problem 7: Application Health Checker

def check_system_health(error_count):
    """
    Function to evaluate application health.
    """
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print("Error Count:", error_count)
    print("System Status:", status)


# Sample Data
check_system_health(7)
```

## Problem 9: E-Learning – Student Attendance Eligibility System

```python
# Problem 9: Attendance Eligibility System

def check_attendance(attendance_list):
    """
    Function to calculate attendance percentage and exam eligibility.
    """
    total_classes = len(attendance_list)
    attended = 0
    
    for status in attendance_list:
        if status == "Present":
            attended += 1
    
    percentage = (attended / total_classes) * 100
    
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)


# Sample Data
attendance_data = ["Present", "Absent", "Present", "Present", "Present"]
check_attendance(attendance_data)
```

## Problem 10: Smart Electricity Bill Analyzer

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    """
    Function to calculate electricity bill and classify usage.
    """
    bill = 0
    
    # Billing slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    
    print("Total Bill:", bill)
    print("Usage Status:", usage)


# Sample Data
calculate_electricity_bill(250)
```
